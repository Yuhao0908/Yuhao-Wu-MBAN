```python
resource_id = '8a89caa9-511c-4568-af89-7f2174b4378c'
limit = 10500
offset = 0

api_url = 'https://data.ontario.ca/api/3/action/datastore_search?resource_id={}&limit={}&offset={}'.format(resource_id,limit,offset)
api_response = requests.get(api_url)
vaccines_data = pd.DataFrame(api_response.json()['result']['records'])
vaccines_data.head()
```

```python
for colname in vaccines_data.columns[2:]:
  vaccines_data[colname] = vaccines_data[colname].fillna(0).astype(int)
```

```python
transformed_data = pd.DataFrame({
    'report_date': vaccines_data['report_date'],
    'new_doses': vaccines_data['previous_day_total_doses_administered'],
    'new_doses_unvaccinated': vaccines_data['previous_day_total_doses_administered'] -
                              (vaccines_data['previous_day_at_least_one'] +
                               vaccines_data['previous_day_fully_vaccinated'] +
                               vaccines_data['previous_day_3doses'].fillna(0)),
    'new_doses_one_vaccine': vaccines_data['previous_day_at_least_one'],
    'new_doses_two_vaccines': vaccines_data['previous_day_fully_vaccinated'],
    'new_doses_three_vaccines': vaccines_data['previous_day_3doses'].fillna(0)
})
```

```python
# query to caculate number of doses administered per quarter for all years available in the dataset
with personal_engine.connect() as connection:
  query = text("""
    SELECT
    EXTRACT(YEAR FROM report_date::timestamp) AS year,
    EXTRACT(QUARTER FROM report_date::timestamp) AS quarter,
    SUM(new_doses) AS total_doses_administered
FROM
    final_exam.covid19
GROUP BY
    EXTRACT(YEAR FROM report_date::timestamp),
    EXTRACT(QUARTER FROM report_date::timestamp)
ORDER BY
    year,
    quarter;

  """)
  query_df = pd.read_sql_query(query, connection)
  display(query_df)
```
